```python
# removing duplicates
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
# cleaning rate column
```

```python
df['rate'].unique()
```

```python
# Removing unnecessary data like NEW, dtype n all and /5
```

```python
def handle_rate(value):
    if(value=='NEW' or value == '-'):
       return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handle_rate)
df['rate'].head()
```

```python
# Handling null values
```

```python
df.rate.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)': 'cost2plates','listed_in(type)': 'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df = df.drop(['listed_in(city)'],axis = 1)
```

```python
df['cost2plates'].unique()
```

```python
# removing , from cost2plates
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['cost2plates'] = df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.head()
```

```python
# cleaning rest column
```

```python
df['rest_type'].unique()
```

```python
rest_type = df['rest_type'].value_counts(ascending = False)
rest_type
```

```python
rest_type_lessthan1000 = rest_type[rest_type <1000]
rest_type_lessthan1000
```

```python
# making rest types less than 1000 in frequency as Other
```

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
# Cleaning Location Column
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]

def handle_location(value):
    if (value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()

                                      
```

```python
# Cleaning cuisines Column
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
